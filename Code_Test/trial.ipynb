```python
class FaceData(Dataset):
    def __init__(self, root_path):
        self.root_path = root_path
        self.files = glob.glob(os.path.join(root_path, '*.jpg'))

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        filename = self.files[idx]

        image = io.imread(filename)
        image = transform.resize(image, (256,256))

        return image
```

```python
facedata = FaceData('../data/raw/part1/')
```

```python
image = facedata[10000]
plt.imshow(image)
```

```python
dataloader = DataLoader(facedata, batch_size=50,
                        shuffle=True, num_workers=4, transform=[])
```
