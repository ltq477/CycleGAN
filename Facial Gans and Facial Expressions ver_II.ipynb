### Creation of a Python class

We will not create a class in order to pull images from the dataset. Unlike the previous precedures, we will not only view the dataset, but assign a class which we may use several times throughout. This class shall make it easier to load images and will be useful when creating the model at a later point.

```python
#Creata a class to pull items from the image folder
class FaceDataset(Dataset):
    def __init__(self, MainFolder, transform=None):
        self.MainFolder = MainFolder
        images = os.listdir(MainFolder)
        self.all_images = glob.glob(os.path.join(MainFolder, '*.jpg'))
        self.transform=transform
        
    def __len__(self):
        return len(self.all_images)
    
    def __getitem__(self, index):
        img_loc = os.path.join(self.MainFolder, self.all_images[index])
        image = io.imread(img_loc)
        if self.transform is not None:
            image=Image.fromarray(image.astype('uint8'), 'RGB')
            image=self.transform(image)
        return image
```

#### Creating Global Variables

We will now assign some global variables which we will be using throughout our model

```python
#Creating Labels
real_label = 1.
fake_label = 0.

#Batch size
batch_size = 128

#number of channels 
nc = 3

#Z latent space / vector
nz = 100

#Generator feature maps
ngf = 64

#Discriminator feature maps
ndf = 64

#number of gpu
ngpu = 1

#beta for optimizer
beta1 = .5
```

```python
def weights(m):
    if isinstance(m, nn.Conv2d):
        nn.init.normal_(m.weight, 0.0, 0.02)
    elif isinstance(m, nn.BatchNorm2d):
        nn.init.normal_(m.weight, 0.0, 0.02)
        nn.init.constant_(m.bias, 0)
```

#### Greate the Generator

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
        
        #layer 1
        nn.ConvTranspose2d(nz, ngf*8, kernel_size=4, stride=1, padding=0, 
                           bias=False),
        nn.BatchNorm2d(ngf*8),
        nn.ReLU(True),
        
        #Layer 2
        nn.ConvTranspose2d(ngf*8, ngf*4, kernel_size=4, stride=2, padding=1,
                           bias=False),
        nn.BatchNorm2d(ngf*4),
        nn.ReLU(True),
        
        #Layer 3
        nn.ConvTranspose2d(ngf*4, ngf*2, kernel_size=4, stride=2, padding=1,
                           bias=False),
        nn.BatchNorm2d(ngf*2),
        nn.ReLU(True),
        
        #Layer 4    
        nn.ConvTranspose2d(ngf*2, ngf, kernel_size=4, stride=2, padding=1,
                           bias=False),
        nn.BatchNorm2d(ngf),
        nn.ReLU(True),
        
        #Layer 5
        nn.ConvTranspose2d(ngf, nc, kernel_size=4, stride=2, padding=1, 
                           bias=False),
        nn.Tanh()
            
        )
        
    def forward(self, input):
        
        return self.main(input)
```

```python
# Create the generator
netG = Generator(ngpu).to(device)
netG.apply(weights)
print(netG)
```

#### Create the Discriminiator

```python
class Discriminator(nn.Module):
    
    def __init__(self, ngpu):
        super(Discriminator, self).__init__()
        self.ngpu = ngpu
        
        self.main = nn.Sequential(
            
            #Layer 1
            nn.Conv2d(nc, ndf, kernel_size=4, stride=2, padding=1,
                     bias=False),
            nn.LeakyReLU(0.2),
            
            #Layer 2
            nn.Conv2d(ndf, ndf*2, kernel_size=4, stride=2, padding=1,
                     bias=False),
            nn.BatchNorm2d(ndf*2),
            nn.LeakyReLU(0.2),
            
            #Layer 3
            nn.Conv2d(ndf*2, ndf*4, kernel_size=4, stride=2, padding=1,
                     bias=False),
            nn.BatchNorm2d(ndf*4),
            nn.LeakyReLU(0.2),
            
            #Layer 4
            nn.Conv2d(ndf*4, ndf*8, kernel_size=4, stride=2, padding=1,
                     bias=False),
            nn.BatchNorm2d(ndf*8),
            nn.LeakyReLU(0.2),
            
            #Layer 5
            nn.Conv2d(ndf*8, 1, kernel_size=4, stride=1, padding=0,
                     bias=False),
            nn.Sigmoid()
        )

    def forward(self, input):
        
        return self.main(input)
```

```python
netD = Discriminator(ngpu).to(device)   
netD.apply(weights)
print(netD)
```

```python
criterion = nn.BCELoss()
noise_f = torch.rand(64, 64, 1, 1, device=None)
```
